- Create an array with shape (4, 3) of: 

a. all zeros 

b. ones 

c. numbers from 0 to 11

- Tabulate the following function: F(x)=2x^2+5, x∈[1,100] with step 1.

- Tabulate the following function: F(x)=e^−x, x∈[−10,10] with step 1.

```python
import numpy as np

a = np.zeros((4, 3))
print(f'all zeros: {a}')

b = np.ones((4, 3))
print(f'ones : {a}')

c = np.arange(12)
print(f'numbers from 0 to 11: {c}')
```

```python
# Tabulate the following function: F(x)=2x^2+5, x∈[1,100] with step 1.
from tabulate import tabulate

x = np.arange(1, 101, 1)
F_x = 2 * x ** 2 + 5

table = np.column_stack((x, F_x))
print(tabulate(table, headers = ["x", "F_x"], tablefmt="github", numalign="center"))
```

```python
# Tabulate the following function: F(x)=e^−x, x∈[−10,10] with step 1.
from tabulate import tabulate

x = np.arange(-10, 11, 1)
F_x = np.exp(-x)

table = np.column_stack((x, F_x))
print(tabulate(table, headers = ["x", "F_x"], tablefmt="github",  numalign="center"))
```

- Import the dataset from this address and assign it to df variable.

- Select only the Team, Yellow Cards and Red Cards columns.

- How many teams participated in the Euro2012?

- Filter teams that scored more than 6 goals

```python
# Import the dataset from this address and assign it to df variable
import pandas as pd

dataset_url = 'https://raw.githubusercontent.com/guipsamora/pandas_exercises/master/02_Filtering_%26_Sorting/Euro12/Euro_2012_stats_TEAM.csv'
df = pd.read_csv(dataset_url)
```

```python
# Select only the Team, Yellow Cards and Red Cards columns.
select = df[['Team','Yellow Cards','Red Cards']]
print(select)
```

```python
# How many teams participated in the Euro2012?
count_teams = df['Team'].nunique()
print(count_teams)
```

```python
# Filter teams that scored more than 6 goals
select = df.loc[df.Goals > 6]
print(select['Team'])
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

penguins = sns.load_dataset("penguins")
```

```python
fig, axs = plt.subplots(ncols = 3)
fig.set_size_inches(16, 5)

sns.histplot(penguins['body_mass_g'], bins = 10, ax = axs[0])
sns.kdeplot(penguins['body_mass_g'], fill=True, bw_adjust=.5, cut=0, ax=axs[1])
_ = sns.ecdfplot(penguins['body_mass_g'], ax = axs[2])
```

```python
_ = sns.regplot( x = 'bill_length_mm',y='body_mass_g', data = penguins)
plt.title("Regression plot")
plt.show()
```

```python
_ = sns.stripplot( x = 'species',y='body_mass_g', data = penguins)
plt.title("Strip plot")
plt.show()
```

```python
_ = sns.catplot(y='sex', x = 'body_mass_g',color="y",order = ["Male", "Female"], data = penguins, kind="swarm")
plt.title("Cat plot")
plt.show()
```

```python
_ = sns.catplot(x='species', y='body_mass_g', hue = 'island', kind = 'box',palette = "muted", data = penguins)
plt.title("Distribution plot")
plt.show()
```

```python
g = sns.catplot(x='sex', y='bill_length_mm',  kind = 'violin',inner = None, data = penguins)
sns.swarmplot(x='sex', y='bill_length_mm', color = 'k', size = 3, data = penguins, ax = g.ax)
plt.title("Combined plots")
plt.show()
```

```python
sns.jointplot(data=penguins,
              x="body_mass_g",
              y="bill_length_mm",
              kind="scatter"
             )

plt.suptitle("Penguins Body Mass", y=1.02)
plt.show()
```
